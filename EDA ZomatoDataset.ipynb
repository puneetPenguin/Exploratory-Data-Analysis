## Missing values

```python
# Checking for null values using sum functionality.
df.isna().sum()
```

We can see 9 missing values in cuisines. Now we have to handle these missing values in some way.

```python
# Checking columns having null values using list comprehension.
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# Checking null values using heatmap.
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

As 9 NaN values are very less in comparison of length of dataframe (9551). So, it is not visible in heatmap.

```python
# Checking the country code data file
df_country = pd.read_excel('Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
# We can combine the dataframes w.r.t country code.
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head(2)
```

```python
# Another way to check datatypes.
final_df.dtypes
```

## Exploring the data

```python
# Checking how many data is from which country
final_df.Country.value_counts()
```

```python
# Separating both columns to plot pie chart for the data.
country_names = final_df.Country.value_counts().index
country_values = final_df.Country.value_counts().values
```

```python
# Pie Chart (Country vs data count) - Top 3 countries.
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
plt.show()
```

**Observation**: Maximum records/ transactions of zomato lies in India.

```python
final_df.columns
```

```python
# Grouping our data according to rating params.
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
ratings
```

**Observation**
1. When rating is between 4.5 to 4.9 --> Excellent.
2. When rating is between 4.0 to 4.4 --> Very Good.
3. When rating is between 3.5 to 3.9 --> Good.
4. When rating is between 2.5 to 3.4 --> Average.
5. When rating is between 1.8 to 2.4 --> Poor.

```python
# Plotting rating and its count on barplot.
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

```python
# Trying to make the graph with including the rating colors
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'], hue='Rating color')
plt.show()
```

Observation:
   1. Not rated count is very high.
   2. Maximum ratings are between 2.5 to 3.4

```python
# Plotting the count plot (It is just checking presence of each colors in the ratings df)
sns.countplot(x='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```
